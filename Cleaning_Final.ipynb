```python
import pandas as pd
import numpy as np

# Loading the merged CSV file
try:
    df = pd.read_csv("All_in_one.csv")
    print("File 'All_in_one.csv' loaded successfully.")
    print(f"Original shape: {df.shape}")
except FileNotFoundError:
    print("Error: 'All_in_one.csv' not found. Please make sure the file is in the correct directory.")
    df = pd.DataFrame() 

if not df.empty:
    # Standardizing District and State Names
    for col in ['District', 'State_x', 'State_y', 'State_left', 'State_right', 'State']:
        if col in df.columns:
            df[col] = df[col].astype(str).str.lower().str.strip().str.replace('&', 'and', regex=False)

    district_mapping = {
        'nicobars': 'andaman and nicobar islands',
        'north and middle andaman': 'andaman and nicobar islands'
    }
    if 'District' in df.columns:
        df['District'] = df['District'].replace(district_mapping)

    # Consolidating Merged Columns
    cols_to_drop = []
    for col in df.columns:
        if col.endswith('_left'):
            base_name = col[:-5]
            right_col = f"{base_name}_right"
            if right_col in df.columns:
                df[base_name] = df[right_col].combine_first(df[col])
                cols_to_drop.extend([col, right_col])
        elif col.endswith('_x'):
            base_name = col[:-2]
            y_col = f"{base_name}_y"
            if y_col in df.columns:
                df[base_name] = df[y_col].combine_first(df[col])
                cols_to_drop.extend([col, y_col])

    df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
    print(f"Consolidated {len(cols_to_drop)} suffixed columns.")

    # Final Cleaning and Type Conversion (Corrected)
    unnamed_cols = [col for col in df.columns if 'Unnamed:' in col]
    df.drop(columns=unnamed_cols, inplace=True, errors='ignore')
    print(f"Dropped {len(unnamed_cols)} 'Unnamed' columns.")

    # Identifying and converting the columns that should be numeric
    cols_to_convert = [col for col in df.columns if col not in ['State', 'District']]

    for col in cols_to_convert:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # Filling missing values with 0 only in these numeric columns
    df[cols_to_convert] = df[cols_to_convert].fillna(0)
    print("Converted columns to numeric and filled missing values with 0.")

    # Final Cleaned DataFrame
    print("\n--- Cleaning Complete! ---")
    print(f"Final cleaned shape: {df.shape}")
    print(df.head())

    df.to_csv("All_data_cleaned.csv", index=False)
    print("\nFully cleaned data saved to 'All_data_cleaned.csv'")
```

```python
import pandas as pd
import numpy as np

# Loading merged CSV file
try:
    df = pd.read_csv("All_data_cleaned.csv")
    print("File 'All_in_one.csv' loaded successfully.")
    print(f"Original shape: {df.shape}")
except FileNotFoundError:
    print("Error: 'All_in_one.csv' not found.")
    df = pd.DataFrame()

if not df.empty:
    # Standardizing District Names
    if 'District' in df.columns:
        df['District'] = df['District'].astype(str).str.lower().str.strip().str.replace('&', 'and', regex=False)
    if 'State' in df.columns:
        df['State'] = df['State'].astype(str).str.lower().str.strip().str.replace('&', 'and', regex=False)

    district_mapping = {
        'ahmadnagar': 'ahmednagar', 'anugul': 'angul', 'baleswar': 'balasore',
        'banas kantha': 'banaskantha', 'bangalore rural': 'bengaluru rural',
        'bangalore urban': 'bengaluru urban', 'baramula': 'baramulla', 'baudh': 'boudh',
        'belgaum': 'belagavi', 'bellary': 'ballari', 'kochbihar': 'cooch behar',
        'debagarh': 'deogarh', 'dhaulpur': 'dholpur', 'east karbi anglong': 'karbi anglong',
        'west karbi anglong': 'karbi anglong', 'jagatsinghapur': 'jagatsinghpur',
        'jajapur': 'jajpur', 'jalor': 'jalore', 'jyotiba phule nagar': 'amroha',
        'kheri': 'lakhimpur kheri', 'mahesana': 'mehsana', 'mumbai city': 'mumbai',
        'mumbai suburban': 'mumbai', 'mysore': 'mysuru', 'narsimhapur': 'narsinghpur',
        'nav sari': 'navsari', 'north and middle andaman': 'andaman and nicobar islands',
        'nicobars': 'andaman and nicobar islands', 'pashchim champaran': 'west champaran',
        'pashchimi singhbhum': 'west singhbhum', 'purbi singhbhum': 'east singhbhum',
        'purw champaran': 'east champaran', 'sabar kantha': 'sabarkantha',
        'sahibzada ajit singh nagar': 'mohali', 'sant kabir nagar': 'sant kabir nagar',
        'seraikela kharsawan': 'saraikela-kharsawan',
        'sri potti sriramulu nellore': 'nellore', 'the dangs': 'dang', 'yadagiri': 'yadgir'
    }
    if 'District' in df.columns:
        df['District'] = df['District'].replace(district_mapping)
        print("Standardized district names.")

    # Grouping by Year/District and Aggregating the Rows

    key_cols = ['Year', 'District']
    if 'State' in df.columns:
        key_cols.append('State')

    numeric_cols = [col for col in df.columns if pd.api.types.is_numeric_dtype(df[col]) and col not in key_cols]

    # Creating aggregation dictionary dynamically
    agg_dict = {col: 'sum' for col in numeric_cols}
    
    # Only adding 'State' to the aggregation if it exists
    if 'State' in df.columns:
        agg_dict['State'] = 'first'

    group_cols = ['Year', 'District']
    df_merged = df.groupby(group_cols).agg(agg_dict).reset_index()
    print(f"Merged duplicate rows. New shape: {df_merged.shape}")

    # Final Cleanup
    final_cols_order = key_cols + sorted([col for col in df_merged.columns if col not in key_cols])
    df_final = df_merged[final_cols_order]

    print("\n--- Cleaning and Merging Complete! ---")
    display(df_final.head())

    df_final.to_csv("All_data_cleaned_and_merged.csv", index=False)
    print("\nFully cleaned and merged data saved to 'All_data_cleaned_and_merged.csv'")
```
